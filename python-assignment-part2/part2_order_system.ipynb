#Task 1 — Explore the Menu

```python
#given menu dictionary

menu = {
    "Paneer Tikka":   {"category": "Starters",  "price": 180.0, "available": True},
    "Chicken Wings":  {"category": "Starters",  "price": 220.0, "available": False},
    "Veg Soup":       {"category": "Starters",  "price": 120.0, "available": True},
    "Butter Chicken": {"category": "Mains",     "price": 320.0, "available": True},
    "Dal Tadka":      {"category": "Mains",     "price": 180.0, "available": True},
    "Veg Biryani":    {"category": "Mains",     "price": 250.0, "available": True},
    "Garlic Naan":    {"category": "Mains",     "price":  40.0, "available": True},
    "Gulab Jamun":    {"category": "Desserts",  "price":  90.0, "available": True},
    "Rasgulla":       {"category": "Desserts",  "price":  80.0, "available": True},
    "Ice Cream":      {"category": "Desserts",  "price": 110.0, "available": False},
}
```

```python
# extract unique categories using set
categories = set(item["category"] for item in menu.values())

for category in categories:  # loop through each category
    print(f"\n===== {category} =====")

    for item, details in menu.items():  # loop through menu items
        if details["category"] == category:

            status = "Available" if details["available"] else "Unavailable"

            # formatted output
            print(f"{item:<15} ₹{details['price']:.2f}   [{status}]")
```

```python
# total number of items
total_items = len(menu)

# total available items
available_items = sum(1 for item in menu.values() if item["available"])

# most expensive item
most_expensive = max(menu.items(), key=lambda x: x[1]["price"])

# all items under ₹150
cheap_items = [(name, data["price"]) for name, data in menu.items() if data["price"] < 150]

print("\nTotal items:", total_items)
print("Available items:", available_items)
print(f"Most expensive item: {most_expensive[0]} (₹{most_expensive[1]['price']})\n")

print("\nItems under ₹150:")
for name, price in cheap_items:
    print(f"{name} - ₹{price}")
```

#Task 2 — Cart Operations

```python
# customer placing order simulation

cart = [] # cart as a list of dictionaries, starting empty

# logic to add item
def add_to_cart(item_name, quantity):

    # check if item exists in menu
    if item_name not in menu:
        print("Item not found in menu\n")
        return

    # check if the item is marked as available
    if not menu[item_name]["available"]:
        print("Item is currently unavailable\n")
        return

    # if the item is already in the cart, increase its quantity instead of adding a duplicate entry
    for item in cart:
        if item["item"] == item_name:
            item["quantity"] += quantity
            print(f"Updated {item_name} quantity to {item['quantity']}")
            return

    # add new item to cart
    cart.append({
        "item": item_name,
        "quantity": quantity,
        "price": menu[item_name]["price"]
    })
    print(f"Added {item_name} x{quantity}")

# logic to remove an item from the cart by name
def remove_from_cart(item_name):
    for item in cart:
        if item["item"] == item_name:
            cart.remove(item)
            print(f"Removed {item_name}")
            return
    print("Item not in cart\n")
```

```python
#simulate the following sequence and print the cart state after each step:
add_to_cart("Paneer Tikka", 2)
print(f"Cart: {cart}\n")

add_to_cart("Gulab Jamun", 1)
print(f"Cart: {cart}\n")

add_to_cart("Paneer Tikka", 1)  # should update
print(f"Cart: {cart}\n")

add_to_cart("Mystery Burger", 1)
add_to_cart("Chicken Wings", 1)

remove_from_cart("Gulab Jamun")
print(f"Cart: {cart}\n")
```

```python
#final order summary

print("\n========== Order Summary ==========")

subtotal = 0

for item in cart:
    total_price = item["quantity"] * item["price"]
    subtotal += total_price

    print(f"{item['item']:<15} x{item['quantity']}   ₹{total_price:.2f}")

gst = subtotal * 0.05
total = subtotal + gst

print("------------------------------------")
print(f"Subtotal:           ₹{subtotal:.2f}")
print(f"GST (5%):           ₹{gst:.2f}")
print(f"Total Payable:      ₹{total:.2f}")
print("====================================")
```

#Task 3 — Inventory Tracker with Deep Copy

```python
import copy
```

```python
# given inventory
inventory = {
    "Paneer Tikka":   {"stock": 10, "reorder_level": 3},
    "Chicken Wings":  {"stock":  8, "reorder_level": 2},
    "Veg Soup":       {"stock": 15, "reorder_level": 5},
    "Butter Chicken": {"stock": 12, "reorder_level": 4},
    "Dal Tadka":      {"stock": 20, "reorder_level": 5},
    "Veg Biryani":    {"stock":  6, "reorder_level": 3},
    "Garlic Naan":    {"stock": 30, "reorder_level": 10},
    "Gulab Jamun":    {"stock":  5, "reorder_level": 2},
    "Rasgulla":       {"stock":  4, "reorder_level": 3},
    "Ice Cream":      {"stock":  7, "reorder_level": 4},
}
```

```python
# deep copy
inventory_backup = copy.deepcopy(inventory)

# prove deep copy
inventory["Paneer Tikka"]["stock"] = 2

print("\nModified Inventory:", inventory["Paneer Tikka"])
print("\nBackup Inventory:", inventory_backup["Paneer Tikka"])

# restore inventory to original state
inventory = copy.deepcopy(inventory_backup)
```

```python
# simulate order fulfilment

# deduct quantities from the final cart in Task 2 from the corresponding items in inventory
for item in cart:
    name = item["item"]
    qty = item["quantity"]

    #if stock for an item is insufficient, print a warning and deduct only what is available (do not go below 0)
    if inventory[name]["stock"] >= qty:
        inventory[name]["stock"] -= qty
        print(f"Deducted {qty} {name}(s) from inventory")
    else:
        print(f"Not enough stock for {name}")
        inventory[name]["stock"] = 0
```

```python
# after deduction, loop through inventory and print a Reorder Alert for every item whose stock is at or below its reorder_level
for item, details in inventory.items():
    if details["stock"] <= details["reorder_level"]:
        print(f"⚠ Reorder Alert: {item} — Only {details['stock']} unit(s) left (reorder level: {details['reorder_level']})")

print("\nFinal stock check (Current vs Backup):")
print(f"Current Paneer Tikka: {inventory['Paneer Tikka']['stock']}")
print(f"Backup  Paneer Tikka: {inventory_backup['Paneer Tikka']['stock']}")
```

#Task 4 — Daily Sales Log Analysis

```python
# sales log
sales_log = {
    "2025-01-01": [
        {"order_id": 1,  "items": ["Paneer Tikka", "Garlic Naan"],          "total": 220.0},
        {"order_id": 2,  "items": ["Gulab Jamun", "Veg Soup"],              "total": 210.0},
        {"order_id": 3,  "items": ["Butter Chicken", "Garlic Naan"],        "total": 360.0},
    ],
    "2025-01-02": [
        {"order_id": 4,  "items": ["Dal Tadka", "Garlic Naan"],             "total": 220.0},
        {"order_id": 5,  "items": ["Veg Biryani", "Gulab Jamun"],           "total": 340.0},
    ],
    "2025-01-03": [
        {"order_id": 6,  "items": ["Paneer Tikka", "Rasgulla"],             "total": 260.0},
        {"order_id": 7,  "items": ["Butter Chicken", "Veg Biryani"],        "total": 570.0},
        {"order_id": 8,  "items": ["Garlic Naan", "Gulab Jamun"],           "total": 130.0},
    ],
    "2025-01-04": [
        {"order_id": 9,  "items": ["Dal Tadka", "Garlic Naan", "Rasgulla"], "total": 300.0},
        {"order_id": 10, "items": ["Paneer Tikka", "Gulab Jamun"],          "total": 270.0},
    ],
}
```

```python
# total revenure per day
daily_revenue = {}
print("Revenue per day:")
for date, orders in sales_log.items():
    total = sum(order["total"] for order in orders)
    daily_revenue[date] = total
    print(f"{date} - ₹{total}")


# best selling day
best_day = max(daily_revenue, key=daily_revenue.get)
print("\nBest selling day:", best_day,"\n")


# logic to find most ordered item
item_count = {}

for orders in sales_log.values():
    for order in orders:
        for item in order["items"]:
            item_count[item] = item_count.get(item, 0) + 1

most_ordered = max(item_count, key=item_count.get)
print("Most ordered item:", most_ordered,"\n")
```

```python
# add new day to sales log
sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 490.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"], "total": 260.0},
]

# reprint revenue per day
print("Updated revenue per day:")
for date, orders in sales_log.items():
    total = sum(order["total"] for order in orders)
    print(f"{date} - ₹{total}")
```

```python
# print a numbered list of all orders across all dates
all_orders = []

for date, orders in sales_log.items():
    for order in orders:
      all_orders.append((date, order))

for idx, (date, order) in enumerate(all_orders, start=1):
    items = ", ".join(order["items"])
    print(f"{idx}. [{date}] Order #{order['order_id']} — ₹{order['total']} — Items: {items}")
```
